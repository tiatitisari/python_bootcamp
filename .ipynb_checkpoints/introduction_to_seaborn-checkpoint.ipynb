```python
# Import Matplotlib and Seaborn
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
np.bool = np.bool_
np.float = np.float_


# Create a scatter plot of absences vs. final grade
sns.scatterplot(x='absences',y='G3',data=student_data, hue='location', hue_order = ['Rural','Urban'])

# Show plot
plt.show()
```

```python
# Import Matplotlib and Seaborn
import matplotlib.pyplot as plt
import seaborn as sns
np.float = np.float_

# Create a dictionary mapping subgroup values to colors
palette_colors = {'Rural': "green", 'Urban': "blue"}

# Create a count plot of school with location subgroups
sns.countplot(x='school',data=student_data,hue='location', palette = palette_colors)



# Display plot
plt.show()
```

```python
# Change to use relplot() instead of scatterplot()
sns.relplot(x="absences", y="G3",
                data=student_data, kind = 'scatter')

# Show plot
plt.show()
```

```python
# Change this scatter plot to arrange the plots in rows instead of columns
sns.relplot(x="absences", y="G3", 
            data=student_data,
            kind="scatter", 
            col="study_time") # arrange in col, if you want to arrange in row use row =  

# Show plot
plt.show()
```

```python
# Adjust to add subplots based on school support
sns.relplot(x="G1", y="G3", 
            data=student_data,
            kind="scatter",col = 'schoolsup', 
            col_order = ['yes','no'])

# Show plot
plt.show()
```

```python
# Adjust further to add subplots based on family support
sns.relplot(x="G1", y="G3", 
            data=student_data,
            kind="scatter", 
            col="schoolsup",
            col_order=["yes", "no"], 
            row = 'famsup', row_order = ['yes','no'])

# Show plot
plt.show()
```

# Line plots: 
two type of relational plots: scatter plots and line plots 
- scatter plots : each plot point is an independent obeservation
- line plots : each plot point represents the same "thing", typically tracked over time: additional accesories>> markers, dashes
  1. multiple observations per x-value, the line graph gonna have confidential interval as shady line. ci represent as mean, but you can change into sd

```python
# Import Matplotlib and Seaborn
import matplotlib.pyplot as plt
import seaborn as sns

# Create line plot
sns.relplot(x='model_year', y= 'mpg', data = mpg, kind = 'line', ci = 'sd')


# Show plot
plt.show()
```

```python
# Import Matplotlib and Seaborn
import matplotlib.pyplot as plt
import seaborn as sns

# Create line plot of model year vs. horsepower
# sns.relplot(x='model_year', y='horsepower', ci = None, data = mpg, kind = 'line')
sns.lineplot(x='model_year', y='horsepower', data = mpg, ci=None)


# Show plot
plt.show()
```

```python
# Import Matplotlib and Seaborn
import matplotlib.pyplot as plt
import seaborn as sns

# Change to create subgroups for country of origin
sns.relplot(x="model_year", y="horsepower", 
            data=mpg, kind="line", 
            ci=None, 
            style = 'origin', 
            hue='origin',
            dashes = False)

# Show plot
plt.show()
```

```python
# Import Matplotlib and Seaborn
import matplotlib.pyplot as plt
import seaborn as sns

# Add markers and make each line have the same style
sns.relplot(x="model_year", y="horsepower", 
            data=mpg, kind="line", 
            ci=None, style="origin", 
            hue="origin", dashes = False, 
markers = True, estimator = np.mean)

# Show plot
plt.show()
```

```python
# Create count plot of internet usage
sns.catplot(x = 'Internet usage', data = survey_data, kind = 'count')


# Show plot
plt.show()
```

```python
# Change the orientation of the plot
sns.catplot(y="Internet usage", data=survey_data,
            kind="count")

# Show plot
plt.show()
```

```python
age_category = []
for i in survey_data['Age']:
    if i < 21:
        age_category.append('Less than 21')
    else:
        age_category.append('21+')

survey_data['Age Category']= age_category 
# # Separate into column subplots based on age category
sns.catplot(y="Internet usage", col = 'Age Category', data=survey_data,
            kind="count")

# # # Show plot
plt.show()
```

```python
survey_data['Interested in Math']= survey_data['Mathematics']>=3 
# survey_data['Interested in Math'] 
# # Create a bar plot of interest in math, separated by gender
sns.catplot(kind = 'bar', x = 'Gender', y='Interested in Math', data = survey_data)


# # Show plot
plt.show()
```

```python
# Create bar plot of average final grade in each study category
sns.catplot(kind='bar', data = student_data, x = 'study_time', y= 'G3')



# Show plot
plt.show()
```

```python
import numpy as np
# List of categories from lowest to highest
category_order = ["<2 hours", 
                  "2 to 5 hours", 
                  "5 to 10 hours", 
                  ">10 hours"]

# Turn off the confidence intervals
sns.catplot(x="study_time", y="G3",
            data=student_data,
            kind="bar",
            order=category_order, 
            ci = None, estimator= sum)

# Show plot
plt.show()
```

# Box Plots 
- shows the distribution of quantitatve data
- see median, spread, skewness and outliers
- facilitates comparisons between groups
- kind = 'box', order, sym --> delete outlier , whis >> if you want to change whisker, by default 1.5 * IQR but extendable to w.0*IQR, show 5th and 95th then whis =[5,95], min and max value = [0,100]
- 

```python
np.int = np.int_
# Specify the category ordering
study_time_order = ["<2 hours", "2 to 5 hours", 
                    "5 to 10 hours", ">10 hours"]

# Create a box plot and set the order of the categories
sns.catplot(x='study_time', y = 'G3', data = student_data,kind = 'box', order = study_time_order, 
        )


# Show plot
plt.show()
```

```python
# Create a box plot with subgroups and omit the outliers
sns.catplot(data = student_data, x = 'internet', y = 'G3', kind = 'box', sym = '', hue = 'location')





# Show plot
plt.show()
```

```python
# Extend the whiskers 0.5*IQR
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box",
            whis=0.5)

# Show plot
plt.show()
```

```python
# Set the whiskers at the min and max values
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box",
            whis=[5, 95])

# Show plot
plt.show()
```

```python
# Set the whiskers at the min and max values
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box",
            whis=[0, 100])

# Show plot
plt.show()
```

# Point Plots 
- show mean of quantitative variable
- vertical lne show 95% ci
- point plots vs line plots
  1. both show: mean of quantitative variable, 95% confidence intervals for the mean
  2. differences: line plot has quantitative variable(usually time) on x-axis
  3. point plot has categorical variable on x-axis
- kind = 'point', join = False , estimator = median , capsize= 0.2, ci = None 

```python
# Remove the lines joining the points
sns.catplot(x="famrel", y="absences",
			data=student_data,
            kind="point",
            capsize=0.2, 
            join = False)
            
# Show plot
plt.show()
```

```python
# Import median function from numpy
from numpy import median

# Plot the median number of absences instead of the mean
sns.catplot(x="romantic", y="absences",
			data=student_data,
            kind="point",
            hue="school",
            ci=None, estimator = median)

# Show plot
plt.show()
```

# Changing plot style and color 
- Reasons to change style
  1. personal preference
  2. improve readability
  3. guide interpretation
- figure 'style' includes background and axes
- preset options : 'white', 'dark', 'whitegrid', 'darkgrid', 'ticks'
- sns.set_style()
- figure 'palette' changes the color of the main elements of the plot >> sns.set_palette()
  1. diverging palettes : 'RdBu', 'PRGn', 'RdBu_r'. 'PRGn_r'
  2. sequential palettes: 'Greys', 'Blues', 'PuRd', 'GnBu'
  3. Hex or color code
- figure 'context' changes the scale of the plot elements and labels >> sns.set_context()
  1. smallest to largest: 'paper', 'notebook', 'talk', 'poster'
  2. default is paper 

```python
# Change the color palette to "RdBu"
sns.set_style("whitegrid")
sns.set_palette("RdBu")

# Create a count plot of survey responses
# category_order = ["Never", "Rarely", "Sometimes", 
#                   "Often", "Always"]

sns.catplot(x="Parents' advice", 
            data=survey_data, 
            kind="count")

# Show plot
plt.show()

# survey_data.columns
```

```python
# Change the context to "poster"
sns.set_context("talk")

# Create bar plot
sns.catplot(x="Siblings", y="Loneliness",
            data=survey_data, kind="bar")

# Show plot
plt.show()
# survey_data
```

```python
# Set the style to "darkgrid"
sns.set_style('darkgrid')

# Set a custom color palette
sns.set_palette(["#39A7D0","#36ADA4"])

# Create the box plot of age distribution by gender
sns.catplot(x="Gender", y="Age", 
            data=survey_data, kind="box")

# Show plot
plt.show()
```

# Adding titles and labels : Part 1 
- Seaborn plots create two different types of objects: FacetGrid and AxesSubplot
- type(g) g as documentation
- FacetGrid vs AxesSubplot objects:
  1. FacetGrid : plot types relplot(), catplot() >> characteristics can create subplots
     - g.fig.suptitle('New Title', y = 1.03) >> y to adjust height
     - g.set_title("This is {col_name}")
     - g.set (xlabel = "New X label",
                ylabel = "New Y Label")
     - plt.xticks(rotation = 90) >> rotate the axis name  
  3. AxesSubplot: plot types scatterplot(), countplot(), etc >> only creates a single plot
     - g.set_title('New Title', y=1.03)
     - g.set (xlabel = "New X label",
                ylabel = "New Y Label")

```python
# Create scatter plot
g = sns.relplot(x="weight", 
                y="horsepower", 
                data=mpg,
                kind="scatter")

# Identify plot type
g.fig.suptitle('Car Weight vs. Horsepower', y = 1.03)

# Print type
# print(type_of_g)
```

```python
# Create line plot
g = sns.lineplot(x="model_year", y="mpg", 
                 data=mpg,
                 hue="origin", 
                 ci = None)

# Add a title "Average MPG Over Time"
g.set_title("Average MPG Over Time")

# Add x-axis and y-axis labels
g.set(xlabel = 'Car Model Year', ylabel = 'Average MPG')

# Show plot
plt.show()
```

```python
# Create point plot
sns.catplot(x="origin", 
            y="acceleration", 
            data=mpg, 
            kind="point", 
            join=False, 
            capsize=0.1)

# Rotate x-tick labels
plt.xticks(rotation = 90)

# Show plot
plt.show()
```

```python
# Set palette to "Blues"
sns.set_palette('Blues')

survey_data['Interested in Pets']=survey_data['Pets']>=3
# Adjust to add subgroups based on "Interested in Pets"
g = sns.catplot(x="Gender",
                y="Age", data=survey_data, 
                kind="box", hue='Interested in Pets')

# Set title to "Age of Those Interested in Pets vs. Not"
g.fig.suptitle("Age of Those Interested in Pets vs. Not")

# Show plot
plt.show()
```

```python
# Set the figure style to "dark"
sns.set_style('dark')
survey_data['Likes Techno']=survey_data['Techno']>=3
# Adjust to add subplots per gender
g = sns.catplot(x="Village - town", y="Likes Techno", 
                data=survey_data, kind="bar",
                col='Gender')

# Add title and axis labels
g.fig.suptitle("Percentage of Young People Who Like Techno", y=1.02)
g.set(xlabel="Location of Residence", 
       ylabel="% Who Like Techno")

# Show plot
plt.show()
```
